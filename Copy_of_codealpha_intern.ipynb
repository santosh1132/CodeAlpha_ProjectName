TASK 3 - DATA VISUALIZATION

1. Created a Bar Chart to visualize survival count.
2. Created a Pie Chart to show survival percentage.
3. Created a Histogram to understand Passenger ID distribution.
4. Created a Box Plot to identify outliers.
5. Visualizations helped understand the dataset clearly.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
df['Survived'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title("Survival Percentage")
plt.ylabel("")
plt.show()
```

```python
plt.hist(df['PassengerId'], bins=20)
plt.title("Passenger ID Distribution")
plt.xlabel("Passenger ID")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(y=df['PassengerId'])
plt.title("Passenger ID Box Plot")
plt.show()
```
